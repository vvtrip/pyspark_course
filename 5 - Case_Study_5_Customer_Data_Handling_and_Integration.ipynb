```python
import pandas as pd
df= pd.read_csv('/content/FairDealCustomerData.csv',header=None)
df.head()
```

```python
df.tail()
```

```python
df.rename(columns={0:'lastname',1:'name',2:'not_allowed'}, inplace=True)
```

```python
df
```

```python
names = [obj.split() for obj in df['name']]
names[:5]
```

```python
title = []
fname = []
midname = []

for obj in names:
  title.append(obj[0])
  fname.append(obj[1])
  try:
    midname.append(obj[2])
  except:
    midname.append(obj.append(''))

df.drop(['name'],axis=1,inplace=True)

df['title'] = title
df['fname'] = fname
df['midname'] = midname

df.head()
```

```python
new_cols = ['title', 'fname', 'midname','lastname','not_allowed']
df=df.reindex(columns=new_cols)
df.head()
```
